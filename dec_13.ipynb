```python
text='''#.##..##.
..#.##.#.
##......#
##......#
..#.##.#.
..##..##.
#.#.##.#.

#...##..#
#....#..#
..##..###
#####.##.
#####.##.
..##..###
#....#..#'''
```

```python
def find_symmetry(matrix):
    for i in range(len(matrix) - 1):
        if matrix[i] == matrix[i+1]:
            short_part = min(i+1, len(matrix) - (i+1))
            if matrix[(i+1)-short_part:i+1] == matrix[i+1:(i+1)+short_part][::-1]:
                return i + 1
    return -1
```

```python
# PART 1:
matrices = [[[c for c in row] for row in matrix.split("\n")] for matrix in text.split("\n\n")]
total_sum = 0
for matrix in matrices:
    output = find_symmetry(matrix)
    if output != -1:
        total_sum += 100*output
        print(f"{output}: {total_sum}")
    matrix_t = list(zip(*matrix))
    output = find_symmetry(matrix_t)
    if output != -1:
        total_sum += output
        print(f"{output}: {total_sum}")
print(total_sum)
        
```
