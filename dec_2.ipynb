```python
games: list = text.split("\n")
games_split: list[dict] = []
for game in games:
    split = game.split(":")
    name, grabs = split[0], split[-1] 
    grabs_list: list = grabs.split(";")
    id_in = get_number_from_string(name)
    game_dict = { "game_id": id_in, "grabs": grabs_list}
    games_split.append(game_dict)

total_sum = 0

for game in games_split:
    # red, green, blue
    min_cubes = [0,0,0]
    for grab in game["grabs"]:
        split = grab.split(",")
        for cubes in split:
            number_color = cubes.split(" ")
            number, color = int(number_color[1]), number_color[2]
            if color == "red" and number > min_cubes[0]:
                min_cubes[0] = number
            elif color == "green" and number > min_cubes[1]:
                min_cubes[1] = number
            elif color == "blue" and number > min_cubes[2]:
                min_cubes[2] = number
    total_sum += min_cubes[0]*min_cubes[1]*min_cubes[2]   
    print(f"{game['grabs']}, {min_cubes}, {total_sum}")
    
print(total_sum)
```
