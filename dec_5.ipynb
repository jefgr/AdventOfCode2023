```python
seeds_split = [int(x) for x in seeds.split(" ")]
seed_to_soil_rows = seed_to_soil_map.split("\n")
soil_to_fertilizer_rows = soil_to_fertilizer_map.split("\n")
fertilizer_to_water_rows = fertilizer_to_water_map.split("\n")
water_to_light_rows = water_to_light_map.split("\n")
light_to_temperature_rows = light_to_temperature_map.split("\n")
temperature_to_humidity_rows = temperature_to_humidity_map.split("\n")
humidity_to_location_rows = humidity_to_location_map.split("\n")
map_list = [seed_to_soil_rows, soil_to_fertilizer_rows, fertilizer_to_water_rows, water_to_light_rows, light_to_temperature_rows, temperature_to_humidity_rows, humidity_to_location_rows]
```

```python
def find_in_map(input: int, map: list) -> int:
    for row in map:
        row_split = row.split(" ")
        row_split = [int(x) for x in row_split]
        if row_split[1] + row_split[2] > input >= row_split[1]:
            return row_split[0] + (input - row_split[1])
    return input
```

```python
from math import inf

# PART 1:
lowest_location = inf
for seed in seeds_split:
    result = seed
    for map in map_list:
        result = find_in_map(result, map)
    if result < lowest_location:
        lowest_location = result
    print(f"{seed}: {result}")
print(lowest_location)
```

```python
# PART 2: takes very long
lowest_location = inf
lowest_info = []
for a, b in zip(seeds_split[::2], seeds_split[1::2]):
    for seed in range(a, a+b):
        result = seed
        for map in map_list:
            result = find_in_map(result, map)
        if result < lowest_location:
            lowest_location = result
            lowest_info = [a, b, seed]
        print(f"{lowest_location}, {seed}")
print(f"{lowest_location}")
print(lowest_info)
```
