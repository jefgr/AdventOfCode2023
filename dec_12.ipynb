```python
def is_valid_row(symbols, numbers):
    if group_sizes(symbols) == numbers:
        return True
    return False
```

```python
def group_sizes(row):
    group_sizes_list = []
    current_group_size = 0

    for symbol in row:
        if symbol == '#':
            current_group_size += 1
        elif current_group_size > 0:
            group_sizes_list.append(current_group_size)
            current_group_size = 0

    if current_group_size > 0:
        group_sizes_list.append(current_group_size)

    return group_sizes_list
```

```python
# PART 1:
from itertools import product

rows = text.splitlines()
rows_symbols = []
rows_numbers = []
for row in rows:
    output = row.split(" ")
    rows_symbols.append([c for c in output[0]])
    rows_numbers.append([int(c) for c in output[1].split(",")])
    
total_sum = 0
for row, numbers in zip(rows_symbols, rows_numbers):
    s = 0
    count = row.count("?")
    for comb in product(".#", repeat=count):
        comb_iter = iter(comb)
        new_row = [next(comb_iter) if c =='?' else c for c in row]
        if is_valid_row(new_row, numbers):
            s += 1
    total_sum += s
    print(f"{s}, {total_sum}")
print(f"Total Sum: {total_sum}")
```

```python
from itertools import permutations
from copy import deepcopy

# PART 2: very slow
rows_copy = deepcopy(rows_symbols)
[row.append("?") for row in rows_copy]
unfolding_rows_symbols = [row * 5 for row in rows_copy]
unfolding_rows_symbols = [row[:-1] for row in unfolding_rows_symbols]
unfolding_rows_numbers = [row * 5 for row in rows_numbers]


total_sum = 0
for row, numbers in zip(unfolding_rows_symbols, unfolding_rows_numbers):
    s = 0
    count = row.count("?")
    count_broken_in_row = row.count('#')
    count_total_broken = sum(numbers)
    to_shuffle_symbols = "#" * (count_total_broken-count_broken_in_row) + "." * (count - count_total_broken + count_broken_in_row)
    print(f"{row}: {count}: {count_broken_in_row}: {count_total_broken}: {to_shuffle_symbols}")
    for comb in permutations(to_shuffle_symbols, count):
        comb_iter = iter(comb)
        new_row = [next(comb_iter) if c =='?' else c for c in row]
        if is_valid_row(new_row, numbers):
            s += 1
    total_sum += s
    print(f"{s}, {total_sum}")
print(f"Total Sum: {total_sum}")
```

```python
# Not completely mine
from functools import cache


@cache
def num_solutions(s, sizes, num_done_in_group=0):
    if not s:
        # Is this a solution? Did we handle and close all groups?
        return not sizes and not num_done_in_group
    num_sols = 0
    # If next letter is a "?", we branch
    possible = [".", "#"] if s[0] == "?" else s[0]
    for c in possible:
        if c == "#":
            # Extend current group
            num_sols += num_solutions(s[1:], sizes, num_done_in_group + 1)
        else:
            if num_done_in_group:
                # If we were in a group that can be closed, close it
                if sizes and sizes[0] == num_done_in_group:
                    num_sols += num_solutions(s[1:], sizes[1:])
            else:
                # If we are not in a group, move on to next symbol
                num_sols += num_solutions(s[1:], sizes)
    return num_sols
ws = [l.split() for l in text.split("\n")]
rows_ = [(w1, tuple(map(int, w2.split(",")))) for w1, w2 in ws]
print(sum(num_solutions("?".join([group] * 5) + ".", sizes * 5) for group, sizes in rows_))
```
