```python
# PART 1:
rows = text.split("\n")
rows_split = [x.split(" ") for x in rows]
hands_list = []
for row in rows_split:
    cards = [c for c in row[0]]
    hand_value = 0
    highest_count = 0
    second_count = 0
    card_set = set(cards)
    for card in card_set:
        count = cards.count(card)
        if count > highest_count:
            second_count = highest_count
            highest_count = count
        elif count > second_count:
            second_count = count
    if highest_count == 5:
        hand_value = 7
    elif highest_count == 4:
        hand_value = 6
    elif highest_count == 3:
        if second_count == 2:
            hand_value = 5
        else:
            hand_value = 4
    elif highest_count == 2:
        if second_count == 2:
            hand_value = 3
        else:
            hand_value = 2
    else:
        hand_value = 1
    cards = list(map(lambda x: x.replace('A','14'),cards))
    cards = list(map(lambda x: x.replace('K','13'),cards))
    cards = list(map(lambda x: x.replace('Q','12'),cards))
    cards = list(map(lambda x: x.replace('J','11'),cards))
    cards = list(map(lambda x: x.replace('T','10'),cards))
    cards = [int(c) for c in cards]
    value_list = [hand_value] + cards
    hands_list.append((value_list, int(row[1])))
hands_list.sort(key=lambda x: x[0])
total_sum = 0
for i, hand in enumerate(hands_list):
    total_sum += (i+1) * hand[1]
    print(f"{i+1}, {hand[1]}, {total_sum}")
print(total_sum)
```

```python
# PART 2:
# PART 1:
rows = text.split("\n")
rows_split = [x.split(" ") for x in rows]
hands_list = []
for row in rows_split:
    cards = [c for c in row[0]]
    hand_value = 0
    highest_count = 0
    second_count = 0
    jokers = cards.count("J")
    card_set = set(cards)
    for card in card_set:
        if card == "J":
            continue
        count = cards.count(card)
        if count > highest_count:
            second_count = highest_count
            highest_count = count
        elif count > second_count:
            second_count = count
    highest_count += jokers
    if highest_count == 5:
        hand_value = 7
    elif highest_count == 4:
        hand_value = 6
    elif highest_count == 3:
        if second_count == 2:
            hand_value = 5
        else:
            hand_value = 4
    elif highest_count == 2:
        if second_count == 2:
            hand_value = 3
        else:
            hand_value = 2
    else:
        hand_value = 1
    cards = list(map(lambda x: x.replace('A','14'),cards))
    cards = list(map(lambda x: x.replace('K','13'),cards))
    cards = list(map(lambda x: x.replace('Q','12'),cards))
    cards = list(map(lambda x: x.replace('J','1'),cards))
    cards = list(map(lambda x: x.replace('T','10'),cards))
    cards = [int(c) for c in cards]
    value_list = [hand_value] + cards
    hands_list.append((value_list, int(row[1])))
hands_list.sort(key=lambda x: x[0])
total_sum = 0
for i, hand in enumerate(hands_list):
    total_sum += (i+1) * hand[1]
    print(f"{i+1}, {hand[1]}, {total_sum}")
print(total_sum)
```
