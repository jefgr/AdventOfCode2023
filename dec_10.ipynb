```python
text='''..F7.
.FJ|.
SJ.L7
|F--J
LJ...'''
```

```python
def find_element(matrix, element):
    for i, row in enumerate(matrix):
        for j, value in enumerate(row):
            if value == element:
                return i, j
    return None  # Return None if the element is not found
```

```python
def find_next(previous_loc, current_loc):
    next_loc = 0, 0
    if matrix[current_loc[0]][current_loc[1]] == "|":
        next_loc = current_loc[0] + (current_loc[0] - previous_loc[0]), current_loc[1]
    elif matrix[current_loc[0]][current_loc[1]] == "-":
        next_loc = current_loc[0], current_loc[1] + (current_loc[1] - previous_loc[1])
    elif matrix[current_loc[0]][current_loc[1]] == "L":
        if current_loc[0] == previous_loc[0]:
            next_loc = current_loc[0] - 1, current_loc[1]
        else:
            next_loc = current_loc[0], current_loc[1] + 1
    elif matrix[current_loc[0]][current_loc[1]] == "J":
        if current_loc[0] == previous_loc[0]:
            next_loc = current_loc[0] - 1, current_loc[1]
        else:
            next_loc = current_loc[0], current_loc[1] - 1
    elif matrix[current_loc[0]][current_loc[1]] == "7":
        if current_loc[0] == previous_loc[0]:
            next_loc = current_loc[0] + 1, current_loc[1]
        else:
            next_loc = current_loc[0], current_loc[1] - 1
    elif matrix[current_loc[0]][current_loc[1]] == "F":
        if current_loc[0] == previous_loc[0]:
            next_loc = current_loc[0] + 1, current_loc[1]
        else:
            next_loc = current_loc[0], current_loc[1] + 1
    elif matrix[current_loc[0]][current_loc[1]] == ".":
        raise ValueError
    elif matrix[current_loc[0]][current_loc[1]] == "S":
        print("found S")
        next_loc = -1, -1
    return next_loc
```
