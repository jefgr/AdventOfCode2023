```python
text='''...#......
.......#..
#.........
..........
......#...
.#........
.........#
..........
.......#..
#...#.....'''
```

```python
def manhattan_distance(point1, point2):
    return abs(point1[0] - point2[0]) + abs(point1[1] - point2[1])
```

```python
from itertools import combinations

# PART 1:
rows = text.splitlines()
rows_ = []
for row in rows:
    rows_.append(row)
    if "#" not in row:
        rows_.append(row)

rows = list(zip(*rows_))
rows_ = []
for row in rows:
    rows_.append(row)
    if "#" not in row:
        rows_.append(row)
rows = list(zip(*rows_))
locations = set()
for y in range(len(rows)):
    for x in range(len(rows[0])):
        if rows[y][x]=="#":
            locations.add((x,y))
print(locations)
total_sum = 0
for i, j in combinations(locations, 2):
    total_sum += manhattan_distance(i, j)
print(total_sum)
```

```python
# PART 2:
rows = text.splitlines()
rows_index = [0]
columns_index = [0]
for row in rows:
    if "#" not in row:
        rows_index.append(rows_index[-1] + 1000000)
    else:
        rows_index.append(rows_index[-1] + 1)

rows = list(zip(*rows))
for row in rows:
    if "#" not in row:
        columns_index.append(columns_index[-1] + 1000000)
    else:
        columns_index.append(columns_index[-1] + 1)
rows = list(zip(*rows))
locations = set()
for y in range(len(rows)):
    for x in range(len(rows[0])):
        if rows[y][x]=="#":
            locations.add((columns_index[x], rows_index[y]))

total_sum = 0
for i, j in combinations(locations, 2):
    total_sum += manhattan_distance(i, j)
print(total_sum)
```
