```python
def parse_string_to_dict(input_string):
    split = input_string.split(",")
    split2 = [combination.split("=") for combination in split]
    # Create a dictionary from the matches
    result_dict = {comb[0]: int(comb[1]) for comb in split2}
    
    return result_dict
```

```python
def run_workflow(part, workflow_name):
    workflow = workflows.get(workflow_name)
    end = workflow[-1][0]
    for condition in workflow[:-1]:        
        target = condition[0][0]
        operator = condition[0][1]
        value = int(condition[0][2:])
        cond = False
        if operator == ">":
            cond = part.get(target) > value
        elif operator == "<":
            cond = part.get(target) < value
        if cond and condition[1] == 'A':
            return sum(part.values())
        elif cond and condition[1] == 'R':
            return 0
        elif cond:
            return run_workflow(part, condition[1])
    if end == "A":
        return sum(part.values())
    elif end == "R":
        return 0
    else:
        return run_workflow(part, end)
```
