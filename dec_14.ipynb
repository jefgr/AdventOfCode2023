```python
def print_pretty_matrix(matrix):
    for row in matrix:
        print(" ".join(map(str, row)))
```

```python
# PART 1:
matrix = [[c for c in row] for row in text.splitlines()]
matrix_t = list(zip(*matrix))
new_matrix = []
for row in matrix_t:
    new_row = []
    group = []
    for c in row:
        if c == '#':
            group.sort()
            group.reverse()
            group.extend(c)
            new_row.extend(group)
            group = []
        else:
            group.extend(c)
    group.sort()
    group.reverse()
    new_row.extend(group)
    new_matrix.append(new_row)
    
new_matrix_t = list(zip(*new_matrix))
print_pretty_matrix(new_matrix_t)
total_sum = 0
length = len(new_matrix_t)
for i, row in enumerate(new_matrix_t):
    total_sum += sum([length - i if c == 'O' else 0 for c in row])
print(total_sum)
```

```python
def tilt(start_matrix): # always to the left
    end_matrix = []
    for row in start_matrix:
        new_row_ = []
        group_ = []
        for c in row:
            if c == '#':
                group_.sort(reverse=True)
                group_.extend(c)
                new_row_.extend(group_)
                group_ = []
            else:
                group_.extend(c)
        group_.sort(reverse=True)
        new_row_.extend(group_)
        end_matrix.append(new_row_)
    return end_matrix

def rotate_matrix_90_right(matrix):
    # Use zip(*matrix) to transpose the matrix
    transposed_matrix = list(zip(*matrix))
    # Reverse each row to get the 90-degree right rotation
    rotated_matrix = [list(reversed(row)) for row in transposed_matrix]
    return rotated_matrix

def rotate_matrix_90_left(matrix):
    # Reverse each row to get the 90-degree left rotation
    rotated_matrix = [list(reversed(row)) for row in matrix]
    # Use zip(*rotated_matrix) to transpose the matrix back
    transposed_matrix = list(zip(*rotated_matrix))
    return transposed_matrix

def do_a_cycle(m): # North, West, South, East
    matrix_n = rotate_matrix_90_left(m)
    matrix_n = tilt(matrix_n)
    matrix_w = rotate_matrix_90_right(matrix_n)
    matrix_w = tilt(matrix_w)
    matrix_s = rotate_matrix_90_right(matrix_w)
    matrix_s = tilt(matrix_s)
    matrix_e = rotate_matrix_90_right(matrix_s)
    matrix_e = tilt(matrix_e)
    nm = rotate_matrix_90_right(rotate_matrix_90_right(matrix_e))
    return nm
```
