```python
def move(direction):
    if direction == "U":
        return [-1, 0]
    if direction == "R":
        return [0, +1]
    if direction == "D":
        return [+1, 0]
    if direction == "L":
        return [0, -1]
```

```python
def print_pretty_matrix(matrix):
    for row in matrix:
        print(" ".join(map(str, row)))
```

```python
def flood_fill(screen: list[list[int]], sourcerow: int, sourcecol: int, row: int, col: int, source: int, color: int) -> None:
    # Condition for checking out of bounds
    if sourcerow < 0 or sourcerow >= row or sourcecol < 0 or sourcecol >= col:
        return
 
    if screen[sourcerow][sourcecol] != source:
        return
 
    screen[sourcerow][sourcecol] = color
    flood_fill(screen, sourcerow - 1, sourcecol, row, col, source, color)  # left
    flood_fill(screen, sourcerow + 1, sourcecol, row, col, source, color)  # right
    flood_fill(screen, sourcerow, sourcecol + 1, row, col, source, color)  # top
    flood_fill(screen, sourcerow, sourcecol - 1, row, col, source, color)  # bottom
```
