```python
import re
dictionaries = []
for row in text.split("\n"):
        # Using regular expression to extract variable names
        matches = re.match(r"(\w{3}) = \((\w{3}), (\w{3})\)", row)
        if matches:
            parsed_dict = {'name': matches.group(1), 'L': matches.group(2), 'R': matches.group(3)}
            # Printing the dictionary
            print(parsed_dict)
            dictionaries.append(parsed_dict)
        else:
            print("String does not match the expected format.")
```

```python
# PART 1:
from itertools import cycle

current_dict  = next((item for item in dictionaries if item["name"] == "AAA"))
count = 0
instructions_list = [c for c in instructions if c.isalpha()]
instructions_cycle = cycle(instructions_list)
while current_dict["name"] != "ZZZ":
    direction = next(instructions_cycle)
    print(f"{direction}: {current_dict}")
    current_dict  = next((item for item in dictionaries if item["name"] == current_dict[direction]))
    count += 1
    
print(count)
```

```python
from math import gcd


def lcm_of_list(numbers):
    def lcm(x, y):
        return x * y // gcd(x, y)

    result = 1
    for number in numbers:
        result = lcm(result, number)
    return result
```

```python
# PART 2:
starting_dicts = [d for d in dictionaries if d["name"].endswith("A")]
instructions_list = [c for c in instructions if c.isalpha()]
counts = []
for current_dict in starting_dicts:
    count = 0
    instructions_cycle = cycle(instructions_list)
    while not current_dict["name"].endswith("Z"):
        direction = next(instructions_cycle)
        current_dict  = next((item for item in dictionaries if item["name"] == current_dict[direction]))
        count += 1
    counts.append(count)
lowest_common_multiple = lcm_of_list(counts)
print(lowest_common_multiple)
```
