```python
import re
def replace_words_with_digits(input_string):
    number_words = [
        ('one', 'o1e'),
        ('two', 't2o'),
        ('three', 't3e'),
        ('four', 'f4r'),
        ('five', 'f5e'),
        ('six', 's6x'),
        ('seven', 's7n'),
        ('eight', 'e8t'),
        ('nine', 'n9e')
    ]
    
    # Sort the number words by length in descending order
    number_words.sort(key=lambda x: len(x[0]), reverse=True)
    
    for string, character in number_words:
         input_string = input_string.replace(string, character)

    return input_string
```
