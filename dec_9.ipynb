```python
text='''0 3 6 9 12 15
1 3 6 10 15 21
10 13 16 21 30 45'''
```

```python
def find_differences(original_list):
    new_list = []
    for a, b in zip(original_list, original_list[1:]):
        diff = b - a
        new_list.append(diff)
    return new_list
```

```python
# PART 1:
rows = text.split("\n")
total_sum = 0
for row in rows:
    answer = 0
    row_list = [int(x) for x in row.split(" ")]
    while any(element != 0 for element in row_list):
        answer += row_list[-1]
        row_list = find_differences(row_list)
        print(answer)
    total_sum += answer
    print(f"total sum:{total_sum}, answer: {answer}")
print(total_sum)
```

```python
# PART 2:
rows = text.split("\n")
total_sum = 0
for row in rows:
    answer_list = []
    row_list = [int(x) for x in row.split(" ")]
    while any(element != 0 for element in row_list):
        answer_list.append(row_list[0])
        row_list = find_differences(row_list)
        print(row_list[0])
    answer = 0
    for l in answer_list[::-1]:
        answer = l - answer
    total_sum += answer
    print(f"total sum:{total_sum}, answer: {answer}")
print(f"total_sum: {total_sum}")
```
