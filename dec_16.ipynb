```python
def move(direction):
    if direction == "N":
        return [-1, 0]
    if direction == "E":
        return [0, +1]
    if direction == "S":
        return [+1, 0]
    if direction == "W":
        return [0, -1]
```

```python
def tile_effect(pointer):
    row = pointer[0]
    column = pointer[1]
    direction = pointer[2]
    if matrix[row][column] == "|" and direction in ["E", "W"]:
        direction = "N"
        pointers.append([row, column, "S"])
    elif matrix[row][column] == "-" and direction in ["N", "S"]:
        direction = "E"
        pointers.append([row, column, "W"])
    elif matrix[row][column] == "/":
        if direction == "N":
            direction = "E"
        elif direction == "E":
           direction = "N"
        elif direction == "S":
            direction = "W"
        elif direction == "W":
            direction = "S"
    elif matrix[row][column] == "\\":
        if direction == "N":
            direction = "W"
        elif direction == "W":
           direction = "N"
        elif direction == "S":
            direction = "E"
        elif direction == "E":
            direction = "S"
    return [row, column, direction]
```

```python
def print_pretty_matrix(matrix):
    for row in matrix:
        print(" ".join(map(str, row)))
```

```python
def determine_activation(start_pointer):
    pointers.clear()
    pointers.append(start_pointer)
    activation_matrix = [[0 for i in range(col_count)] for j in range(row_count)]
    activation_matrix_directions = {"N":[[0 for i in range(col_count)] for j in range(row_count)],"E":[[0 for i in range(col_count)] for j in range(row_count)],"S":[[0 for i in range(col_count)] for j in range(row_count)],"W":[[0 for i in range(col_count)] for j in range(row_count)]}
    for pointer in pointers:
        while pointers:
            new_move = move(pointer[2])
            pointer = [pointer[0] + new_move[0], pointer[1] + new_move[1], pointer[2]]
            if pointer[0] >= row_count or pointer[1] >= col_count or pointer[0] < 0 or pointer[1] < 0:
                break
            if activation_matrix_directions.get(pointer[2])[pointer[0]][pointer[1]] == 1:
                break
            activation_matrix[pointer[0]][pointer[1]] = 1
            activation_matrix_directions.get(pointer[2])[pointer[0]][pointer[1]] = 1
            pointer = tile_effect(pointer)
    total_sum = sum([sum(row) for row in activation_matrix])
    print(total_sum)
    return total_sum
```

```python
matrix = [[c for c in row] for row in text.splitlines()]
pointers = []
print_pretty_matrix(matrix)
row_count = len(matrix)
col_count = len(matrix[0])
determine_activation([0, -1, "E"])
```

```python
# PART 2:
scores = []
matrix = [[c for c in row] for row in text.splitlines()]
print_pretty_matrix(matrix)
for i in range(col_count):
    score = determine_activation([-1, i, "S"])
    score2 = determine_activation([row_count, i, "N"])
    scores.append(score)
    scores.append(score2)
for j in range(row_count):
    score = determine_activation([j, -1, "E"])
    score2 = determine_activation([j, col_count, "W"])
    scores.append(score)
    scores.append(score2)
max_sum = max(scores)
print(f"max sum: {max_sum}")
```
